```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("weather_data1.csv")
df
```

```python
new_df = df.replace(['-99999','-88888'],np.nan) # -99999 is replaced by NaN
new_df
```

```python
new_df = df.replace({
    '-99999' : np.nan,
    '-88888' : np.nan,
    'No Event' : 'Sunny'
  })
new_df
```

```python
new_df = df.replace('[A-Za-z]', '',regex=True)
new_df
```

```python
new_df = df.replace({'Temperature' : '[A-Za-z]',
                    'Windspeed' : '[A-Za-z]'
                    }, '',regex=True)
new_df
```
