## Arithmetic Operations

```python
import pandas as pd
```

```python
arr = pd.DataFrame({"A":[1,6,8,4], "B":[5,4,9,2]})
arr
```

```python
# Addition
arr["C"] = arr["A"] + arr["B"]
arr
```

```python
# Subtraction
arr["D"] = arr["A"] - arr["B"]
arr
```

```python
# Multiplication
arr["E"] = arr["A"] * arr["B"]
arr
```

```python
# Division
arr["F"] = arr["A"]/arr["B"]
arr
```

```python
arr1 = pd.DataFrame({"A":[12,24,10,18], "B":[15,20,12,28]})
arr1
```

```python
arr1["Less Than 20"] = arr1["A"] < 20
arr1
```

```python
arr1["More Than 15"] = arr1["B"] > 15
arr1
```
